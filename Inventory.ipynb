```python
# show_details is a function which returns the information about all the details
def show_details():
    global info
    return info

# Initializing the variable
sale = ''

# Loading the inventory from JSON to String
file = open("Records.json","r")
js = file.read()
file.close()

# Converting String Records to JSON/Dictionaries
record = json.loads(js)

# Printing the Inventory MENU which includes the product name,price,quantity remaining
print(f"{'-'*19}Menu{'-'*30}\n")
for key in record.keys():
    print(f"{key}: {record[key]['Name']} | {record[key]['Price']} | {record[key]['Quantity']}")
print(f"{'-'*50}\n")

# Taking user inputs about the purchase
user_name = input(" Enter your name           : ")
user_email = input("Enter your mail id       : ")
user_phone = input("Enter your phone number  : ")
user_product = str(input("Enter product ID   : "))
user_quantity = int(input("Enter Quantity    : "))
bill = int(user_quantity * record[user_product]['Price'])
gst = round((0.18 * bill),2)
total_bill = bill + gst
info =   f'''
          Name                  :       {user_name}
          User Email            :       {user_email}
          User Phone number     :       {user_phone}
          User Product ID       :       {user_product}
          Product Name          :       {record[user_product]["Name"]}
          Total Bill            :       {total_bill}
          Time of transaction   :       {time.ctime()}\n'''

# Applying the condition :
# Here if the we're having equal to or more quantity that user wants
if record[user_product]["Quantity"] >= user_quantity: 
    
    record[user_product]['Quantity'] -= user_quantity #Updating the records i.e. subtracting from actual inventory

    print(f"{'-'*50}\n")
    #displaying the purchase details
    print(f"Name                :  {record[user_product]['Name']}")
    print(f"Price Per Piece(Rs) :  {record[user_product]['Price']}")
    print(f"Quantity Entered    :  {user_quantity}")
    print(f"GST                 :  {gst}")
    print(f"\n{'-'*19}Bill{'-'*30}\n")
    print(f"Total Bill          : {total_bill} Rs")
    print(f"{'-'*50}")
    print(show_details())
    
# If we're not having enough quantity available that user requried then: 
else:
    print(f"{'-'*50}\n")
    print("Sorry, we're not having enough quantity in our inventory.")
    print(f"We're having {record[user_product]['Quantity']} quantity.")
    print(f"{'-'*50}\n")
    
    choice = input("Press Y/y to purhcase:").lower()
    
    if choice == 'y':           #If the user wants to purchase the remaining quantity available
    
        new_user_quantity = int(input("Enter Quantity    : "))
        record[user_product]['Quantity'] -= new_user_quantity   #Updating the inventory
        #Displaying the Purchase Details
        print(f"Name                :  {record[user_product]['Name']}")
        print(f"Price (Rs)          :  {record[user_product]['Price']}")
        print(f"Quantity Entered    :  {new_user_quantity}")
        print(f"GST                 :  {gst}")
        print(f"\n{'-'*19}Bill{'-'*30}")
        
        bill = int(new_user_quantity) * int(record[user_product]['Price'])
        gst = 0.18*bill
        total_bill = gst + bill
        
        print(f"Total Bill          : {total_bill} Rs")
        print(show_details()) 
        print(f"{'-'*50}")
    
    else:           # If the user entered anything other than Y or y then come out from the condition
        print("Thank You!")    
    
# Converting the dictionary to string
js = json.dumps(record)

# Saving the json file
file = open("Records.json","w")
file.write(js)
file.close()

print(f"{'-'*50}")
print(f"\tThanks for your order, Inventory Updated!")
print(f"{'-'*50}")

# Saving the records in sales file
sale = open("sales.txt","a")
sale.write(info)
sale.close()
```
